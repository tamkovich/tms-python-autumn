# decorators

```python
def log_params(real_sum):
    def fake_sum(a, b):
        print(f"Входные данные: {a}, {b}")
        return real_sum(a, b)
    return fake_sum

def sum_(a, b):
    return a + b
```

```python
sum_
```

```python
sum_(9, 80)
```

```python
sum_ = log_params(sum_)
```

```python
sum_
```

```python
sum_(9, 80)
```

```python
@log_params
def mult_(a, b):
    return a * b
```

```python
mult_(2, 3)
```

# Задание 9.06

Написать декоратор, который будет выводить время выполнения функции

```python
from datetime import datetime
time = datetime.now()
```

```python
time
```

```python
def time_checker(real_func):
    def fake_func(*args):
        print(f"Входные данные: {args}")
        return real_func(*args)
    return fake_func
```

```python
datetime_now = time_checker(datetime.now)
```

```python
datetime_now()
```
