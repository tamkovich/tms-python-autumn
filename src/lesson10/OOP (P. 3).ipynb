### @classmethod

````python
class Pet:
    VOICE = None
    
    @classmethod
    def voice(cls: 'Pet'):
        print(cls.VOICE)
        


class Cat(Pet):
    VOICE = "Meow"
````

### Абстрактный класс

Абстрактный класс - класс, экземпляр которого нельзя создать.

````python
from abc import ABC, abstractmethod

class AbstractDuck(ABC):
    @abstractmethod
    def noise(self):
        pass


class BrazilDuck(AbstractDuck):
    def noise(self):
        print('Кря Кря' )
````

````python
a = AbstractDuck() # ERROR
````

````python
b = BrazilDuck()
b.noise()
````

````python
class ToyDuck(AbstractDuck):
    def noise(self):
        print("I am toy Duck")
````

````python
t = ToyDuck()
````

````python
t = ToyDuck()
t.noise()
````

### Интерфейсы

Интерфейсы - класс, определяющий методы и их сигнатуру, который **Должны** быть переопределены в дочерних классах. Интерфейс - договор

````python
class MyInterface(ABC):
    @abstractmethod
    def do_a(self, arg1):
        raise NotImplemented

    @abstractmethod
    def do_b(self, arg1, arg2):
        raise NotImplemented
````

````python
class MyClass(MyInterface):
    def do_a(self, arg1):
        print(arg1)
    
    def do_b(self, arg1, arg2):
        print(arg1, arg2)
````

### Mixins

Миксины инкапсулируют поведение которое может быть использовано в классах.

````python
class LoginRequiredMixin:
    pass
````

````python
class EngineMixin:
    def run(self):
        print(self.a)

        
class TransmissionMixin:
    def change_gear(self, arg1, arg2):
        print(self.b)


class Car(EngineMixin, TransmissionMixin):
    def __init__(self, a, b):
        self.a = a
        self.b = b
````

````python
eng = EngineMixin()
````

````python
eng.run()
````

````python
obj = Car(10, 20)
obj.run()
obj.change_gear(1, 3)
````
