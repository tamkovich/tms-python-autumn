# Pandas Practice / Numpy Practice

```python
import sklearn
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(
    data=np.c_[iris['data'], iris['target']],
    columns= iris['feature_names'] + ['target'],
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['petal length (cm)', 'sepal length (cm)', 'sepal width (cm)', 'petal width (cm)']]
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
import matplotlib as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(prediction, y_test)
```

#  DecisionTreeClassifier

```python
for depth in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
    model = DecisionTreeClassifier(max_depth=depth, random_state=1)
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print(f"Accuracy for depth = {depth} :", accuracy_score(prediction, y_test))
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
print(f"Accuracy:", accuracy_score(prediction, y_test))
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
rounded_predictions = np.round(prediction)
```

```python
rounded_predictions = rounded_predictions.astype('int')
```

```python
rounded_predictions
```

```python
y_test.values
```

```python
print(f"Accuracy:", accuracy_score(rounded_predictions, y_test))
```
